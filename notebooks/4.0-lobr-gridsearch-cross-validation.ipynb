```python
from sklearn.datasets import load_iris
```

```python
data = load_iris(as_frame=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# stratify  se pasa la variable que estamos interesados para que tome valores TODO
X_train, X_test, y_train, y_test= train_test_split(data.data,data.target,test_size=0.2,shuffle=True, random_state=42, stratify=data.target)
```

```python
from sklearn.linear_model import SGDClassifier
```

```python
model = SGDClassifier(loss='log_loss',random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
SGDClassifier?
```

```python
parameters = {
    'alpha':[0.0001, 0.001,0.01],
      'max_iter':[10, 10,100]}
```

```python
gridsearch= GridSearchCV(model,parameters)

```

```python
gridsearch.fit(X_train,y_train)
```

```python
# estos son los mejores parametros
gridsearch.best_params_
```

```python
gridsearch.cv_results_
```

```python
# equivalente a {'alpha': 0.01, 'max_iter': 100}
model = SGDClassifier(alpha=0.01,max_iter=100,random_state=42)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
